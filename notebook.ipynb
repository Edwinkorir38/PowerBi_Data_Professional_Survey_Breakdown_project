# Data Professional Survey Breakdown – Portfolio Summary

This dashboard was designed and analyzed using Power BI Desktop, showcasing my ability to transform raw survey data into meaningful business insights.

* **Survey Overview:** The dataset includes responses from 630 data professionals with an average age of 29.87.

* **Global Participation:** A country-level treemap highlights the distribution of respondents across regions such as Canada, India, and the UK.

* **Salary Insights:** A comparative bar chart displays the average salary by job title, showing that roles like Data Scientist and Data Architect rank among the highest paid.

* **Technical Skills:** Python is the dominant programming language according to the “Favorite Programming Language” chart.

* **Industry Entry Difficulty:** A donut chart captures perceived difficulty breaking into the data field, with a large portion indicating it was somewhat difficult.

* **Satisfaction Metrics:** Gauge visuals reflect respondents’ happiness with work-life balance (5.74/10) and salary (4.27/10).

This project demonstrates my skills in data cleaning, modeling, visualization, and storytelling using Power BI Desktop, turning survey data into actionable and visually engaging insights.
